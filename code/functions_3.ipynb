```python
try:
    from PIL import Image
except ImportError:
    import Image
import pytesseract
import numpy as np
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import math
import os
from tqdm import tqdm
```

```python
def checkformat(s):
    n = len(s)
    news = '';
    for i in range(n):
        news = news + s[i]
        if s[i] == '°':
            break;
    pos1 = i
    count1 = 0;
    for i in range(pos1+1,n):
        if (s[i] >= '0') and (s[i] <= '9'):
            count1 = count1 + 1
            if (count1 <= 2):
                news = news + s[i]
            else:
                break
    pos1 = i
    news = news + "'"
    count2 = 0;
    for i in range(pos1,n):
        if (s[i] >= '0') and (s[i] <= '9'):
            count2 = count2 + 1
            if (count2 <= 2):
                news = news + s[i]
            else:
                break
    news = news + "."
    pos1 = i
    count3 = 0;
    for i in range(pos1,n):
        if (s[i] >= '0') and (s[i] <= '9'):
            count3 = count3 + 1
            if (count3 <= 2):
                news = news + s[i]
            else:
                break
    news = news + '"'
    return news
```

```python
def direction_and_height(path):
    lena = mpimg.imread(path)
    
    lenb = lena[298:312,130:376, :]
    
    plt.imsave('direction_height.png', lenb)
    
    pytesseract.pytesseract.tesseract_cmd ='E:\\Tesseract-OCR\\tesseract'
    result = pytesseract.image_to_string('direction_height.png')
    
    result = result.replace(',','.')
    result = result.replace('..','.')
    result = result.replace('”','"')
    
    for i in range(len(result)):
        if ((result[i] >= '0') and (result[i] <= '9')) or (result[i] == '-'):
            break
    pos1_start = i
    for i in range(pos1_start,len(result)):
        if result[i] == '"':
            break
    pos1_end = i
    direction = result[pos1_start:pos1_end + 1]
    direction = direction.replace(' ','')
    direction = checkformat(direction)
    
    for i in range(pos1_end,len(result)):
        if ((result[i] >= '0') and (result[i] <= '9')) or (result[i] == '-'):
            break
    pos2_start = i
    for i in range(pos2_start,len(result)):
        if result[i] == '"':
            break
    pos2_end = i
    height = result[pos2_start:pos2_end + 1]
    height = height.replace(' ','')
    height = checkformat(height)
    return direction , height
```

```python
def checkformat_phase(s):
    n = len(s)
    news = ''
    count1 = 0
    for i in range(n):
        if (s[i] >= '0') and (s[i] <= '9'):
            count1 = count1 + 1
            if (count1 <= 1):
                news = news + s[i]
            else:
                break
    pos1 = i
    news = news + '.'
    count2 = 0
    for i in range(pos1,n):
        if (s[i] >= '0') and (s[i] <= '9'):
            count2 = count2 + 1
            if (count1 <= 3):
                news = news + s[i]
            else:
                break
    return news
```

```python
def phase(path):
    lena = mpimg.imread(path)
    lenb = lena[338:353,155:196,:]
    
#     tmp = np.hstack((lenb, lenb, lenb))
#     plt.imsave('phase.png',tmp)
#     plt.imshow(tmp)
    
    plt.imsave('phase.png',lenb)
    pytesseract.pytesseract.tesseract_cmd ='E:\\Tesseract-OCR\\tesseract'
    result = pytesseract.image_to_string(Image.open('phase.png'))
    result = result.replace(',','.')
    result = result.replace('..','.')
    result = result.replace(' ','')
    result = checkformat_phase(result)
    return result
```

```python
def checkdate(s):
    n = len(s)
    news = '';
    count1 = 0;
    for i in range(n):
        if (s[i] >= '0') and (s[i] <= '9'):
            count1 = count1 + 1
            if (count1 <= 4):
                news = news + s[i]
            else:
                break
    pos1 = i
    news = news + "."
    count2 = 0;
    for i in range(pos1,n):
        if (s[i] >= '0') and (s[i] <= '9'):
            count2 = count2 + 1
            if (count2 <= 2):
                news = news + s[i]
            else:
                break
    news = news + "."
    pos1 = i
    count3 = 0;
    for i in range(pos1,n):
        if (s[i] >= '0') and (s[i] <= '9'):
            count3 = count3 + 1
            if (count3 <= 2):
                news = news + s[i]
            else:
                break
    return news
```

```python
def checktime(s):
    n = len(s)
    news = '';
    count1 = 0;
    for i in range(n):
        if (s[i] >= '0') and (s[i] <= '9'):
            count1 = count1 + 1
            if (count1 <= 2):
                news = news + s[i]
            else:
                break
    pos1 = i
    news = news + "h"
    count2 = 0;
    for i in range(pos1,n):
        if (s[i] >= '0') and (s[i] <= '9'):
            count2 = count2 + 1
            if (count2 <= 2):
                news = news + s[i]
            else:
                break
    news = news + "m"
    pos1 = i
    count3 = 0;
    for i in range(pos1,n):
        if (s[i] >= '0') and (s[i] <= '9'):
            count3 = count3 + 1
            if (count3 <= 2):
                news = news + s[i]
            else:
                break
    news = news + '.'
    pos1 = i
    count4 = 0;
    for i in range(pos1,n):
        if (s[i] >= '0') and (s[i] <= '9'):
            count4 = count4 + 1
            if (count4 <= 2):
                news = news + s[i]
            else:
                break
    news = news + 's'
    return news
```

```python
def date(path):
    lena = mpimg.imread(path)
    lenb = lena[1416:1432,36:203,:]
    plt.imsave('date.png',lenb)
    pytesseract.pytesseract.tesseract_cmd ='E:\\Tesseract-OCR\\tesseract'
    result = pytesseract.image_to_string(Image.open('date.png'))
    result = result.replace(',','.')
    result = result.replace('..','.')
    result = result.replace('”','"')
    try:
        day, time = result.split(' ')
        day = checkdate(day)
        time = checktime(time)
    except ValueError:
        day, time = (0, 0)
    return day , time
```

```python
path = 'Moon\\'
f = open('Moon.dat', 'w+')
for img in tqdm(os.listdir(path)):
    day, time = date(path + img)
    result = phase(path + img)
    direction, height = direction_and_height(path + img)
    
    try:
        year = day.split('.')[0]
        month = day.split('.')[1]
        d = day.split('.')[2]
    except AttributeError:
        year = 'N/A'
        month = 'N/A'
        d = 'N/A'
    except IndexError:
        year = 'N/A'
        month = 'N/A'
        d = 'N/A'
    
    try:
        hour = time.split('h')[0]
        m = time.split('h')[1].split('m')[0][1] + time.split('h')[1].split('m')[1][0]
        s = time.split('h')[1].split('m')[1].split('.')[0][1] + time.split('h')[1].split('m')[1].split('.')[1][0]
    except AttributeError:
        hour = 'N/A'
        m = 'N/A'
        s = 'N/A'
    except IndexError:
        hour = 'N/A'
        m = 'N/A'
        s = 'N/A'
    
    try:
        degree_1 = direction.split('°')[0]
        minute_1 = direction.split('°')[1].split('\'')[0]
        second_1 = direction.split('°')[1].split('\'')[1].split('"')[0]
    except IndexError:
        degree_1 = 'N/A'
        minute_1 = 'N/A'
        second_1 = 'N/A'
    
    try:
        degree_2 = height.split('°')[0]
        minute_2 = height.split('°')[1].split('\'')[0]
        second_2 = height.split('°')[1].split('\'')[1].split('"')[0]
    except IndexError:
        degree_2 = 'N/A'
        minute_2 = 'N/A'
        second_2 = 'N/A'
    
    try:
        f.write('{}\t{}\t{}\t{}\t{}\t{}\t{}\t{}\t{}\t{}\t{}\t{}\t{}\n'.format(year, month, d, hour, m, s, \
                                                          degree_1, minute_1, second_1, degree_2, minute_2, second_2, result))
    except UnicodeEncodeError:
        year = 'N/A'
        month = 'N/A'
        d = 'N/A'
        hour = 'N/A'
        m = 'N/A'
        s = 'N/A'
        degree_1 = 'N/A'
        minute_1 = 'N/A'
        second_1 = 'N/A'
        degree_2 = 'N/A'
        minute_2 = 'N/A'
        second_2 = 'N/A'
        f.write('{}\t{}\t{}\t{}\t{}\t{}\t{}\t{}\t{}\t{}\t{}\t{}\t{}\n'.format(year, month, d, hour, m, s, \
                                                          degree_1, minute_1, second_1, degree_2, minute_2, second_2, result))
f.close()
```
